```python
# WoE function for discrete unordered variables
def woe_ordered_continous(x, discrete_variabe_name, nonfraud_fraud_variable_x):
    x = pd.concat([x[discrete_variabe_name], nonfraud_fraud_variable_x], axis = 1)
    x = pd.concat([x.groupby(x.columns.values[0], as_index = False)[x.columns.values[1]].count(),
                    x.groupby(x.columns.values[0], as_index = False)[x.columns.values[1]].mean()], axis = 1)
    x = x.iloc[:, [0, 1, 3]]
    x.columns = [x.columns.values[0], 'n_obs', 'prop_nonfraud']
    x['prop_n_obs'] = x['n_obs'] / x['n_obs'].sum()
    x['n_fraud'] = x['prop_nonfraud'] * x['n_obs']
    x['n_nonfraud'] = (1 - x['prop_nonfraud']) * x['n_obs']
    x['prop_n_nonfraud'] = x['n_nonfraud'] / x['n_nonfraud'].sum()
    x['prop_n_fraud'] = x['n_fraud'] / x['n_fraud'].sum()
    x['WoE'] = np.log(x['prop_n_nonfraud'] / x['prop_n_fraud'])
    #x = x.sort_values(['WoE'])
    #x = x.reset_index(drop = True)
    x['diff_prop_nonfraud'] = x['prop_nonfraud'].diff().abs()
    x['diff_WoE'] = x['WoE'].diff().abs()
    x['IV'] = (x['prop_n_nonfraud'] - x['prop_n_fraud']) * x['WoE']
    # x['IV'] = x['IV'].fillna(0).sum()
    return x
```

```python
# Below we define a function that takes 2 arguments: a dataframe and a number.
# The number parameter has a default value of 0.
# This means that if we call the function and omit the number parameter, it will be executed with it having a value of 0.
# The function displays a graph.
def plot_by_woe(x_WoE, rotation_of_x_axis_labels = 0):
    y = np.array(x_WoE.iloc[:, 0].apply(str))
    # Turns the values of the column with index 0 to strings, makes an array from these strings, and passes it to variable x.
    z = x_WoE['WoE']
    # Selects a column with label 'WoE' and passes it to variable y.
    plt.figure(figsize=(18, 6))
    # Sets the graph size to width 18 x height 6.
    plt.plot(y, z, marker = 'o', linestyle = '--', color = 'k')
    # Plots the datapoints with coordiantes variable x on the x-axis and variable y on the y-axis.
    # Sets the marker for each datapoint to a circle, the style line between the points to dashed, and the color to black.
    plt.xlabel(x_WoE.columns[0])
    # Names the x-axis with the name of the column with index 0.
    plt.ylabel('Weight of Evidence')
    # Names the y-axis 'Weight of Evidence'.
    plt.title(str('Weight of Evidence by ' + x_WoE.columns[0]))
    # Names the grapth 'Weight of Evidence by ' the name of the column with index 0.
    plt.xticks(rotation = rotation_of_x_axis_labels)
    # Rotates the labels of the x-axis a predefined number of degrees.
```
